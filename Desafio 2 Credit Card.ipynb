```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statistics

#Leo el archivo
cc = pd.read_csv(r'C:\Users\albil\OneDrive\Escritorio\CoderHouse\DS\CC\application_record.csv')

#Filtro por estado de trabajo + sexo para reducir cantidad de columnas
m_pension = cc.loc[(cc['NAME_INCOME_TYPE'] == 'Pensioner') & (cc['CODE_GENDER'] == 'M')]
h_higheducation = cc.loc[(cc['NAME_EDUCATION_TYPE'] == 'Higher education') & (cc['CODE_GENDER'] == 'M')]
m_stateservant = cc.loc[(cc['NAME_INCOME_TYPE'] == 'State servant')]
stateservant = cc.loc[(cc['NAME_INCOME_TYPE'] == 'State servant')]
mujeres = cc.loc[(cc['CODE_GENDER'] == 'F') & (cc['NAME_EDUCATION_TYPE'] == 'Higher education')]
hombres = cc.loc[(cc['CODE_GENDER'] == 'M')]
cc
```

```python
#Grafico 1 de Matplotlib

fig, ax = plt.subplots(figsize=(8,4))
x=m_pension["NAME_FAMILY_STATUS"]
y=m_pension["CNT_CHILDREN"]
ax.bar(x, y)
ax.set_title('Cantidad de hijos por estado civil de mujeres pensionadas')
ax.set_ylabel('Cantidad de hijos')
ax.set_xlabel('Estado civil')
```

```python
#Grafico 2 de matplotlib. Relacion sueldo - vivienda de mujeres con alta educaciÃ³n

fig, ax = plt.subplots(figsize=(8,4))
x=mujeres["NAME_HOUSING_TYPE"]
y=mujeres["AMT_INCOME_TOTAL"]
ax.bar(x, y)
ax.set_title('Relacion sueldo - vivienda de mujeres con alta educaciÃ³n')
ax.set_ylabel('Sueldos')
ax.set_xlabel('Vivienda')
ax.tick_params(labelsize=7)
```

```python
#Grafico 3 de matplotlib. Cantidad de mujeres vs cantidad de hombres evaluados

plt.figure(figsize=[4, 4]) 
plt.hist(cc["CODE_GENDER"])
plt.title('AnÃ¡lisis de genero')
plt.xlabel('GÃ©nero')
plt.ylabel('Cantidad de personas')
```

```python
#Grafico 1 de Seaborn. Histograma con el conteo de los tipos de trabajo

b=sns.histplot(data=cc, x="NAME_INCOME_TYPE")
fig=plt.figure(figsize=(5,5))
b.tick_params(labelsize=7)
```

```python
#Grafico 2 de Seaborn. Grafico de barras con ingresos por educacion y genero
sns.barplot(x="CODE_GENDER", y="AMT_INCOME_TOTAL", hue="NAME_EDUCATION_TYPE", data=cc)
```

```python
#Grafico 3 de Seaborn. Grafico de barras con ingresos por tipo de trabajo y genero
ax=sns.barplot(x="CODE_GENDER", y="AMT_INCOME_TOTAL", hue="NAME_INCOME_TYPE", data=cc)
ax.set(xlabel='Genero', ylabel='Ingresos', title='Ingresos por categoria laboral')
```

```python
#Grafico 4 de Seaborn. Grafico de lineas con ingresos por categoria laboral

ax = sns.lineplot(data=cc, x="NAME_INCOME_TYPE", y="AMT_INCOME_TOTAL")
ax.set(xlabel='Categoria laboral', ylabel='Ingresos', title='Ingresos por categoria laboral')
ax.tick_params(labelsize=8)
```

```python
#Grafico 5 de Seaborn.Grafico de lineas con ingresos por educacion

ax2 = sns.lineplot(data=cc, x="NAME_EDUCATION_TYPE", y="AMT_INCOME_TOTAL")
ax2.set(xlabel='Educacion', ylabel='Ingresos', title='Ingresos por educacion')
ax2.tick_params(labelsize=6)
```

```python
#Grafico 6 de Seaborn. Grafico de distribucion de tamaÃ±o de familia por estado civil

plt.figure()
# Figure -level
ax = sns.displot(data=cc, x='CNT_FAM_MEMBERS', kind='kde', hue='NAME_FAMILY_STATUS', fill=True)
ax.set(xlabel='TamaÃ±o de familia', ylabel='Densidad', title='TamaÃ±o de la familia por estado civil')
```
